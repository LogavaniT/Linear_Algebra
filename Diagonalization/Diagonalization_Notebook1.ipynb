# Solution

**T:R^2->R^2**
T is a function that maps vectors from two-dimensional real coordinate spaces to same two-dimensional real coordinate space

**T(x)=Dx** \
where D is a diagonal matrix and x is a basis vector \
e1=[[1],[0]] and e2=[[0],[1]] are basis vector

**Basis Vector:** A basis vector along with other basis vectors together constitutes a basis(foundation) for all the vectors within the vector space

**Step1:** Plot e1 and e1 vectors \
**Step2:** Transform e1 and e2 using Diagonal matrix D \
**Step3:** Take x as [[1],[1]] and transform using same D


```python
import numpy as np
import matplotlib.pyplot as plt

e1=np.array([[1],[0]])
e2=np.array([[0],[1]])
D=np.array([[3,0],[0,6]])
```

```python
#define a function to find D square
def matrix_multiplication(m1,m2):
    m1_rows=m1.shape[0]
    m1_cols=m1.shape[1]
    m2_rows=m2.shape[0]
    m2_cols=m2.shape[1]
    result=np.zeros([m1_rows,m2_cols])

    if m1_cols!=m2_rows:
        print("multiplication is not possible")

    for i in range(m1_rows):
        for j in range(m1_cols):
            for k in range(m2_rows):
                result[i][j]+=m1[i][k]*m2[k][j]
    return result
    
Dsquare=matrix_multiplication(D,D)
print(Dsquare)
```

Square of diagonal matrix is equal to the square of it's diagonal elements

```python
#define a function to plot vectors e1 and e2
def plot_basis_vectors(e1, e2):

    # Create a new figure and axis
    fig, ax = plt.subplots()

    # Plot the first basis vector e1
    ax.arrow(0, 0, e1[0,0], e1[1,0], head_width=0.1, head_length=0.1, fc='r', ec='r',linewidth=2, label='e1')

    # Plot the second basis vector e2
    ax.arrow(0, 0, e2[0,0], e2[1,0], head_width=0.1, head_length=0.1, fc='g', ec='g',linewidth=2, label='e2')

    # Set plot limits and labels
    ax.set_xlim([-0.5, 3])
    ax.set_ylim([-0.5, 3])
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    # Add grid and legend
    ax.grid()
    ax.legend()

    # Show plot
    plt.title('Basis Vectors')
    plt.axhline(0, color='black',linewidth=0.1)
    plt.axvline(0, color='black',linewidth=0.1)
    plt.show()

# Plot the basis vectors
plot_basis_vectors(e1, e2)
```

```python
#define a function for T(x)
def T_of_x(D_matrix,vector):
    rows=D_matrix.shape[0]
    cols=D_matrix.shape[1]
    vrows=vector.shape[0]
    if cols != vrows:
        print("multiplication is not possible")
    result=np.zeros([rows,1])
  
    #perform matrix & vector multiplication
    for i in range(rows):
        for j in range(cols):
            result[i] += D_matrix[i][j]*vector[j]
    return result
```

```python
transformed_e1=T_of_x(D,e1)
transformed_e2=T_of_x(D,e2)
print(transformed_e1)
print(transformed_e2)

#define a function to plot vectors e1 and e2
def plot_transformed_vectors(e1, e2,transformed_e1,transformed_e2):

    # Create a new figure and axis
    fig, ax = plt.subplots()

    # Plot the first basis vector e1
    ax.arrow(0, 0, e1[0,0], e1[1,0], head_width=0.1, head_length=0.1, fc='r', ec='r',linewidth=2, label='e1')

    # Plot the second basis vector e2
    ax.arrow(0, 0, e2[0,0], e2[1,0], head_width=0.1, head_length=0.1, fc='g', ec='g',linewidth=2, label='e2')

    # Plot the transformed e1 vector
    ax.arrow(0, 0, transformed_e1[0,0], transformed_e1[1,0], head_width=0.1, head_length=0.1, fc='b', ec='b',linewidth=1, label='e2')
    
    # Plot the transformed e2 vector
    ax.arrow(0, 0, transformed_e2[0,0], transformed_e2[1,0], head_width=0.1, head_length=0.1, fc='y', ec='y',linewidth=1, label='e2')

    # Set plot limits and labels
    ax.set_xlim([-0.5, 8])
    ax.set_ylim([-0.5, 8])
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    # Add grid and legend
    ax.grid()
    ax.legend()

    # Show plot
    plt.title('Transformed vectors')
    plt.axhline(0, color='black',linewidth=0.1)
    plt.axvline(0, color='black',linewidth=0.1)
    plt.show()

# Plot the basis vectors
plot_transformed_vectors(e1, e2,transformed_e1,transformed_e2)
```

```python
X=np.array([[1],[1]])
transformed_X=T_of_x(D,X)
print(transformed_X)
```

```python
#define a function to plot vectors e1 and e2
def plot_x(X, transformed_X):

    # Create a new figure and axis
    fig, ax = plt.subplots()

    # Plot the first basis vector e1
    ax.arrow(0, 0, X[0,0], X[1,0], head_width=0.1, head_length=0.1, fc='r', ec='r',linewidth=2, label='X')

    # Plot the second basis vector e2
    ax.arrow(0, 0, transformed_X[0,0], transformed_X[1,0], head_width=0.1, head_length=0.1, fc='g', ec='g',linewidth=2, label='Transformed X')

    # Set plot limits and labels
    ax.set_xlim([-0.5, 8])
    ax.set_ylim([-0.5, 8])
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    # Add grid and legend
    ax.grid()
    ax.legend()

    # Show plot
    plt.title('X and Transformed X')
    plt.axhline(0, color='black',linewidth=0.1)
    plt.axvline(0, color='black',linewidth=0.1)
    plt.show()

plot_x(X, transformed_X)
```
