# Solution

**P is invertible.Therefore, the columns of P are linearly independent.Why?** \
For a square matrix P, the matrix is invertible if and only if the determinant is equal to zero.If the columns of P are linearly dependent, then determinant would be zero, making that P is non invertible. Since 
P is invertible, its determinant is non-zero, implying that the columns are linearly independent.


**B is new basis in R^2. Why?** \
If {b1,b2} are linearly independent and can describe every vector in R^2, then they form a basis for R^2

```python
import numpy as np
import matplotlib.pyplot as plt

A=np.array([[4,2],[1,5]])
print("A",A)
P=np.array([[2,1],[-1,1]])
print("P",P)
inv_P=np.array([[1/3,-1/3],[1/3,2/3]])
print("inv_P",inv_P)
b1=np.array([[2],[-1]])
print("b1",b1)
b2=np.array([[1],[1]])
print("b2",b2)
x=np.array([[2],[1]])
print("x",x)
D=np.array([[3,0],[0,6]])
print("D",D)
```

```python
x_in_b_basis=np.dot(inv_P,x)
scaled_x_b_basis=np.dot(D,x_in_b_basis)
x_standard_basis=np.dot(P,scaled_x_b_basis)
print(x_standard_basis)
```

```python
fig,ax=plt.subplots(figsize=(8,8))

scaling_factor = 10
ax.plot([-scaling_factor * b1[0], scaling_factor * b1[0]], [-scaling_factor * b1[1], scaling_factor * b1[1]], color='gray', linestyle='-', linewidth=1.3, label='Basis B')
ax.plot([-scaling_factor * b2[0], scaling_factor * b2[0]], [-scaling_factor * b2[1], scaling_factor * b2[1]], color='gray', linestyle='-', linewidth=1.3)

ax.quiver(0,0,b1[0],b1[1],angles='xy',scale_units='xy',scale=1,color='red')
ax.quiver(0,0,b2[0],b2[1],angles='xy',scale_units='xy',scale=1,color='red')

plt.text(b1[0]-1,b1[1]-0.5,'Basis b1=[2,-1]',fontsize=8,color='black')
plt.text(b2[0]-5,b2[1]+0.1,'Basis b2=[1,1]',fontsize=8,color='black')

plt.plot(x[0], x[1],'bo',markersize=3)
plt.plot(x_standard_basis[0],x_standard_basis[1],'bo',markersize=3)

projection = b1 * x_in_b_basis[0]
ax.plot([projection[0], x[0]], [projection[1], x[1]], color='black', linestyle='--', linewidth=0.8)
projection = b2 * x_in_b_basis[1]
ax.plot([projection[0], x[0]], [projection[1], x[1]], color='black', linestyle='--', linewidth=0.8)

projection = b1 * scaled_x_b_basis[0]
ax.plot([projection[0],x_standard_basis[0]], [projection[1],x_standard_basis[1]], color='green', linestyle='--', linewidth=1)
projection = b2 * scaled_x_b_basis[1]
ax.plot([projection[0],x_standard_basis[0]], [projection[1],x_standard_basis[1]], color='green', linestyle='--', linewidth=1)

plt.text(x[0] + 0.7, x[1] + 0.1, 'x = [2,1] \nB co-ordinates are [1/3,4/3]', fontsize=8)
plt.text(x_standard_basis[0] + 0.7, x_standard_basis[1] + 0.1, 'x_transformed = [10,7] \nB co-ordinates are [1,8]', fontsize=8)

x_ticks = [-5, 0, 5, 10, 15]
ax.set_xlim(-5, 10)
ax.set_ylim(-5, 10)
ax.set_xticks(x_ticks)
y_ticks=[-5,-3,-1,1,3,5,7,9]
ax.set_yticks(y_ticks)

plt.grid(linestyle='--', linewidth=0.9)
plt.legend()


pro1 = b1 * x_in_b_basis[0]
pro2 = b2 * x_in_b_basis[1]
pro3 = b1 * scaled_x_b_basis[0]
pro4 = b2 * scaled_x_b_basis[1]

x_poly = [pro2[0],pro4[0],x_standard_basis[0],pro3[0],pro1[0],x[0]]
y_poly = [pro2[1],pro4[1],x_standard_basis[1],pro3[1],pro1[1],x[1]]
plt.show()
```
