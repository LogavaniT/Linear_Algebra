```python
# To find EigenValues and EigenVectors of the given Matrix
# To prove that u and v are EigenVectors of the Matrix Visually
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[1, 6],
              [5, 2]])

# Define the eigenvectors u and v of the matrix A
u = np.array([[6],[ -5]]) 
v = np.array([[3], [-2]]) 

#Reshape the vectors u and v to 1D vectors
u = np.squeeze(u)
v = np.squeeze(v)
# Compute matrix-vector product AU and AV using dot product
Au = np.dot(A, u)
Av = np.dot(A, v)
```

```python
# Create a new figure for visualization
plt.figure(figsize=(8, 6)) #set figure width and height

# Plot the original EigenVectors u and v
#set x,y co-ordinates,label and color for the vector u
plt.plot([0, u[0]], [0, u[1]], color='blue', label='Original u') 
#set x,y co-ordinates,label and color for the vector v
plt.plot([0, v[0]], [0, v[1]], color='orange', label='Original v') 

# Plot the transformed vectors Au and Av
#set x,y co-ordinates,label,style and color for the vector Au
plt.plot([0, Au[0]], [0, Au[1]], color='green', linestyle='dotted', label='A * u') 
#set x,y co-ordinates,label,style and color for the vector Av
plt.plot([0, Av[0]], [0, Av[1]], color='red', linestyle='dotted', label='A * v') 

# Set x,y axes limits and labels
plt.xlim(-5, 10)
plt.ylim(-8, 6)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.legend()
plt.title('Verification of Eigenvectors')

# Show plot for original vectors vs transformed vectors
plt.show()
```
