```python
import numpy as np
import matplotlib.pyplot as plt

# Define the matrix A
A = np.array([[1, 6],
              [5, 2]])

# Define the eigenvectors u and v
u = np.array([[6],[ -5]]) 
v = np.array([[3], [-2]]) 

# Compute matrix-vector product of AU and AV
Au = np.dot(A, u)
Av = np.dot(A, v)
```

```python
# Plot the vectors
fig, ax = plt.subplots(figsize=(8, 6))

# Plot original vectors and their transformations
ax.arrow(0, 0, u[0], u[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue', label='Original u')
ax.arrow(0, 0, v[0], v[1], head_width=0.1, head_length=0.1, fc='orange', ec='orange', label='Original v')
ax.arrow(0, 0, Au[0], Au[1], head_width=0.1, head_length=0.1, fc='green', ec='green', linestyle='dotted', label='A * u')
ax.arrow(0, 0, Av[0], Av[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue', linestyle='dotted', label='A * v')

# Set plot limits and labels
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Eigenvectors and Their Transformations')
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5)
ax.legend()

# Show plot
plt.show()
```
