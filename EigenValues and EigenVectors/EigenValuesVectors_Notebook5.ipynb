# Solution

**X vector represents the probability of landing on a certain page**

**Step1:** Define a function to multiply matrix and the vector X to compute the probability of landing on each page

**Step2:** Define x1 vector

**Step3:** Compute matrix multiplication of matrix and x1 vector over 100 iterations

**Step4:** Repeat the same for x2 vector

```python
#define the matrix
matrix=[[0,     0,  1,0.5],
        [0.3333,0,  0,0],
        [0.3333,0.5,0,0.5],
        [0.3333,0.5,0,0]]
```

```python
#define a function for the multiplication of matrices
def matrix_mul(matrix,vector):
    result=[]
    for i in range(len(matrix)):
        a=0
        for j in range(len(matrix[0])):
            a += matrix[i][j]*vector[j]
        result.append(a)
    return result
```

```python
#define vectors x1 and x2
x1=[1,0,0,0] #from page 1 


#starting from page 1 moving to next pages depending on the matrix 
#iterating the same process over 100 times
for i in range(100):
    x1=matrix_mul(matrix,x1)
print(x1)

x2=[0,0,1,0] #from page 3
#starting from page 1 moving to next pages depending on the matrix 
#iterating the same process over 100 times
for i in range(100):
    x2=matrix_mul(matrix,x2)
print(x2)
```

The vectors x1 and x2 have almost same probability values of moving to other pages in the network when starting from page 1 and page 3. 

Hence, X is a **Steady-State Vector**
